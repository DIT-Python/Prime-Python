# 코드 13-3 : str_op.py 모듈의 함수들
## str_op.py

```
# 문자열을 대문자 문자열로 만들어서 반환하기
def upper(s):    
  return s.upper()
# 문자열을 소문자 문자열로 만들어서 반환하기
def lower(s):    
  return s.lower()
```
