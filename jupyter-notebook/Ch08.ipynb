# 코드 8-4 : choose_fruit() 함수를 가진 module_fruit.py 스크립트 파일
### module_fruit.py

```python
import random

s = '오늘 먹을 과일을 골라볼까?'

def choose_fruit():
    fruits = ['apple', 'orange', 'banana', 'melon']
    f = random.choice(fruits)
    return f
```

# 코드 8-9 : 몫과 나머지를 반환하는 함수 quot_remain()을 가진 파일
## quot_remain.py

```python
def quot_remain(m, n):
    q = m // n
    r = m % n  
    return q, r
```

# 코드 8-10 : 사용자 정의 모듈인 quot_remain을 import하여 몫과 나머지 구하기
## use_quot_remain.py

```python
import quot_remain

x = 22
y = 4
q, r = quot_remain.quot_remain(x, y)
print('{} / {}의 몫은 {}, 나머지는 {}'.format(x, y, q, r))
```
