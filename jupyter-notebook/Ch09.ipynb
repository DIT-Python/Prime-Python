# 코드 9-8 : try-except-else-finally 구문의 사용법
## divide_func.py

```python
def divide(x, y):
    try:
        result = x / y
    except ZeroDivisionError:
        print("0으로 나누는 오류발생")
    else:
        print("결과 :", result)
    finally:
        print("수행완료")

print('divide(100, 2) 함수호출 :')
divide(100, 2)
print('divide(100, 0) 함수호출 :')
divide(100, 0)
```

# 코드 9-18 : 사용자로부터 입력받은 파일을 번호를 붙여서 출력하기
## file_numbering1.py

```python
fname = input('입력할 파일의 이름 : ')
f = open(fname, 'r')  # 파일 열기
n = 1                 # 줄 번호를 출력하기 위한 변수 
l = f.readline()
while l :      # 읽어들일 줄이 있으면 반복 수행함
    print('{:3d}: {}'.format(n,l), end='') # 줄 번호와 내용을 출력
    n += 1                                 # 줄 번호를 증가시킴
    l = f.readline()

f.close()
```

# 코드 9-19 : try-except 문을 사용한 파일 입출력
## file_numbering3.py

```python
import sys

fname = input('입력할 파일의 이름 : ')
try :
    f = open(fname, 'r', encoding='UTF8')  # 파일 열기
except IOError:
    print('Could not read file: ', fname)
    sys.exit()
except:
    print("Unexpected error:", sys.exc_info()[0])
    sys.exit()

n = 1
l = f.readline()
while l :
    print('{:3d}: {}'.format(n, l), end='')
    n += 1
    l = f.readline()

f.close()
```

# 코드 9-24 : try-except 문을 사용한 파일 입출력
## file_numbering4.py

```python
import sys

fname = input('입력할 파일의 이름 : ')
try:
    with open(fname, 'r', encoding='UTF8') as f: # 파일 열기
        n = 1
        l = f.readline()
        while l :
            print('{:3d}: {}'.format(n, l), end='')
            n += 1
            l = f.readline()
except IOError:
    print('Could not read file: ', fname)
    sys.exit()
except:
    print("Unexpected error:", sys.exc_info()[0])
    sys.exit()
```
