### minus_filter_func.py

```python
def minus_func(n):   # n이 음수이면 True, 그렇지 않으면 False를 반환
    if n < 0 :
        return True
    else:
        return False

n_list = [-30, 45, -5, -90, 20, 53, 77, -36]
minus_list = []     # 음수를 저장할 리스트
for n in filter(minus_func, n_list):    
    minus_list.append(n)
print('음수 리스트 :', minus_list)
```

# 코드 12-25 : 리스트와 튜플, 정수 값을 반복자 객체로 변환시키기

```python
# 리스트가 iterable 객체인가 검사
try:
    l = [10, 20, 30]
    iterator = iter(l)
except TypeError:
    print('list는 iterable 객체가 아닙니다.')
else:
    print('list는 iterable 객체입니다.')

# 튜플이 iterable 객체인가 검사
try:
    t = ('홍길동', 22, 79.7)
    iterator = iter(t)
except TypeError:
    print('tuple은 iterable 객체가 아닙니다.')
else:
    print('tuple은 iterable 객체입니다.')

# 정수형이 iterable 객체인가 검사
try:
    n = 100
    iterator = iter(n)
except TypeError:
    print('n은 iterable 객체가 아닙니다.')
else:
    print('n은 iterable 객체입니다.')
```

# 코드 12-26 : OddCounter 클래스의 정의와 객체 생성
## oddcounter_with_next.py

```python
class OddCounter:
    ''' 1부터 증가하는 홀수를 반환하는 클래스 '''
    def __init__(self, n = 1):
        self.n = n
  
    def __iter__(self):  # 반복자는 __iter__() 함수를 가져야 함
        return self

    def __next__(self):  # 반복자는 __next__() 함수를 가져야 함
        t = self.n
        self.n += 2
        return t
  
my_counter = OddCounter()
print(next(my_counter))
print(my_counter.__next__())
print(my_counter.__next__())
print(my_counter.__next__())
```

# 코드 12-27 : OddCounter 클래스의 정의와 10보다 작은 홀수를 출력하는 방법
## oddcounter_with_for.py

```python
class OddCounter:
    def __init__(self, n = 1):
        self.n = n
  
    def __iter__(self):
        return self

    def __next__(self):
        t = self.n
        self.n += 2
        return t

my_counter = OddCounter()
for x in my_counter:
    if x > 20:
        break
    print(x, end = ' ')
```

# 코드 12-24: Counter 클래스와 StopIteration 예외 생성기능
## oddcounter_with_raise.py

```python
class OddCounter:
    def __init__(self, n = 1):
        self.n = n
  
    def __iter__(self):
        return self

    def __next__(self):
        if self.n < 20:
            t = self.n
            self.n += 2
            return t
        raise StopIteration

my_counter = OddCounter()
for x in my_counter:
    print(x, end = ' ')
```

# 코드 12-25 : 제너레이터의 예제
## generator_with_yield.py

```python
def create_gen():
    alist = range(1, 4)
    for n in alist:
        yield n

agen = create_gen()
print(agen)
for x in agen:
    print(x)
```
